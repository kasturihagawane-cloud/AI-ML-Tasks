```python
import pandas as pd

# Create the dataset
data = {
    'Hours': [1, 2, 2, 4, 5, 6, 5, 4, 6, 9],
    'Score': [35, 45, 48, 52, 60, 68, 75, 82, 88, 95]
}

df = pd.DataFrame(data)
print("Dataset created successfully!")
print(df)
```

```python
# 1. View first 5 rows
print("First 5 rows:")
print(df.head())

# 2. Check shape (rows, columns)
print("\nShape:", df.shape)

# 3. Check data types
print("\nData types:")
print(df.dtypes)

# 4. Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# 5. Basic statistics
print("\nStatistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours'], df['Score'], color='blue', marker='o')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
plt.grid(True)
plt.show()
```
